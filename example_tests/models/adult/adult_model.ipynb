```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('./adult.csv')
```

```python
df.head()
```

```python
model = LogisticRegression().fit(df[['age','educational-num', 'hours-per-week']], df['income'])
```

```python
model.score(df[['age','educational-num', 'hours-per-week']], df['income'])
```

```python
df['out'] = model.predict(df[['age','educational-num', 'hours-per-week']])
```

```python
df.to_csv('adult_pred.csv')
```
